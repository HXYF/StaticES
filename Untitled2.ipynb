```python
import pickle

# 加载 item_clusters.pkl 文件
with open('item_clusters_item.pkl', 'rb') as f:
    item_to_cluster = pickle.load(f)

# 查看前 10 个物品的类别
print("前 10 个物品及其类别：")
for item_id, cluster_id in list(item_to_cluster.items())[:100]:
    print(f"物品 ID: {item_id}, 类别: {cluster_id}")
for item_id, cluster_id in list(item_to_cluster.items())[100:]:
    print(f"物品 ID: {item_id}, 类别: {cluster_id}")
# 查看类别分布统计
from collections import Counter
cluster_counts = Counter(item_to_cluster.values())
print("\n类别分布统计：")
for cluster_id, count in cluster_counts.items():
    print(f"类别 {cluster_id}: {count} 个物品")
```
